```python
import torch
```

```python

cap = cv2.VideoCapture(r"C:\Users\Asmaa\OneDrive\Desktop\IMG_7793.jpeg")

if not cap.isOpened():
    print("فشل في تحميل الفيديو")
else:
    model = torch.hub.load("ultralytics/yolov5", "yolov5s", pretrained=True)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        results = model(frame)
        frame = results.render()[0]
        cv2.imshow("hello", frame)
        if cv2.waitKey(0) == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()
```
